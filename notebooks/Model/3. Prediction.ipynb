```python
from os.path    \
    import (
        join, 
        isdir
)

from os         \
    import (
        makedirs, 
        listdir
)

import cv2

from ultralytics    \
    import YOLO

from PIL import Image 
from os import environ
```

```python
environ['YOLO_VERBOSE'] = 'False'
```

```python
IMAGE_WIDTH_SIZE: int = 640
IMAGE_HEIGHT_SIZE: int = 640

IMAGE_DIMENSIONS: tuple = (
    IMAGE_HEIGHT_SIZE, 
    IMAGE_WIDTH_SIZE
)

LOCATION_OF_MODEL: str = '/mnt/c/Users/Kentv/Desktop/Spectre/Model/#2/weights/best.pt'
LOCATION_OF_PREDICTION_FILES: str = '/mnt/c/Users/Kentv/Desktop/Spectre/Test Data/Videos/TMOD'

LOCATION_TO_SAVE_FILES: str = '/tmp/ModelOutput'
DELETE_SAVE_FILES_ON_FINISH: bool = True
```

```python
if not isdir(
    LOCATION_TO_SAVE_FILES
):
    makedirs(
        LOCATION_TO_SAVE_FILES
    )
```

```python
if isdir(LOCATION_TO_SAVE_FILES):
    %rm {LOCATION_TO_SAVE_FILES} -R
    
    makedirs(
        LOCATION_TO_SAVE_FILES
    )
```

```python
location_of_model: str = LOCATION_OF_MODEL 
Model = None

def getModel():
    global Model
    return Model

def setModel(
    value
):
    global Model
    Model = value

model = YOLO(
    LOCATION_OF_MODEL, 
    verbose=False
)

setModel(
    model
)
```

```python
found_random_files: list = listdir(
    LOCATION_OF_PREDICTION_FILES
)

for idx in range(
    len(
        found_random_files
    )
):
    found_random_files[idx] = join(
        LOCATION_OF_PREDICTION_FILES, 
        found_random_files[idx]
    )

from random import SystemRandom

generator = SystemRandom()

LOCATION_OF_PREDICTION_FILE = found_random_files[
    generator.randint(
        0, 
        int(
            len(
                found_random_files
            ) 
            - 
            1
        )
    )
]
```

```python
counter: int = 0
capture = cv2.VideoCapture(
    LOCATION_OF_PREDICTION_FILE
)

while capture.isOpened():
    (success, frame) = capture.read()
    
    if success:
        counter = counter + 1
        results = model(
            frame,
            verbose=False    
        )

        annotated_frame = results[0].plot()
        
        color_coverted = cv2.cvtColor(
            annotated_frame, 
            cv2.COLOR_BGR2RGB
        )

        image = Image.fromarray(
            color_coverted
        )
          
        image.save(
            join(
                LOCATION_TO_SAVE_FILES, 
                str(
                    str(counter) 
                    + 
                    '.jpg'
                )
            )
        )
    else:
        break
```

```python
if DELETE_SAVE_FILES_ON_FINISH:
    %rm {LOCATION_TO_SAVE_FILES} -R
```
