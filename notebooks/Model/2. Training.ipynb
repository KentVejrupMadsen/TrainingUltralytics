```python
ZERO:   int = 0
ONE:    int = 1
TWO:    int = 2
THREE:  int = 3
FOUR:   int = 4
FIVE:   int = 5

HUNDRED: int = 100
```

```python
from os.path        \
    import (
        join
)
```

```python
IMAGE_WIDTH_SIZE: int = 640
IMAGE_HEIGHT_SIZE: int = 640

IMAGE_DIMENSIONS: tuple = (
    IMAGE_HEIGHT_SIZE, 
    IMAGE_WIDTH_SIZE
)
```

```python
EPOCHS:  int    = 11
BATCHES: int    = 9

IOU:            int     = 95
PATIENCE:       int     = 0
DROPOUT:        float   = 0.10
CONFIDENCE:     float   = 0.95
BOX_PARAMETER:  float   = 5.0

COS_LR: bool   = True
```

```python
MOSAIC_PARAMETER: float = 0.65
MOSAIC_END_AT: int = -ONE
```

```python
CURRENT_NUMBER_OF_TRAINING_ITERATION: int = ONE
NUMBER_OF_TIMES_TO_ITERATE_TRAINING: int = ONE

SEED = None

IS_TO_RESUME: bool = False
HAS_RUN: bool = False

LOAD_DATASET_WORKERS: int = 20
LOAD_STREAM_BUFFER: bool = False

DEVICES_TO_USE: list = [
    ZERO
]
```

```python
HOME_LOCATION: str = '/home/ai'
PROJECT_NAME: str = 'Spectre'
JOB_TYPE: str = 'training'

DATASET_LOCATION: str = join(
    HOME_LOCATION,
    'dataset.yaml'
)

LOAD_MODEL_WEIGHTS_FROM: str = str(
    '/mnt/c/Users/Kentv/Desktop/Spectre/runs/detect/train' 
    + 
    str(29)
    + 
    '/weights/best.pt'
)
```

```python
def calculateIOU() -> float:
    global IOU, HUNDRED

    return float(
        IOU
        /
        HUNDRED
    )
```

```python
CONFIGURATION: dict = {
    "input": {
        "width": IMAGE_WIDTH_SIZE,
        "height": IMAGE_HEIGHT_SIZE
    },
    "model":
    {
        "from": LOAD_MODEL_WEIGHTS_FROM
    },
    "job": 
    {
        "type": JOB_TYPE,
        "dataset": 
        {
            "from": DATASET_LOCATION
        },
        "attributes": 
        {
            "epochs": EPOCHS,
            "batches": 2,
            "iterations": {
                "current": CURRENT_NUMBER_OF_TRAINING_ITERATION,
                "in_total": NUMBER_OF_TIMES_TO_ITERATE_TRAINING
            },
            "data_loaders_in_total": LOAD_DATASET_WORKERS,
            "data_loader_stream": LOAD_STREAM_BUFFER,
            "devices_to_use": DEVICES_TO_USE
        },
        "resuming": IS_TO_RESUME,

    },
    "seed": SEED
}
```

```python
from math           \
    import (
        floor
)

from random         \
    import (
        SystemRandom
)

from os             \
    import (
        environ
)

from logging import ( 
    getLogger, 
    ERROR
)
```

```python
from ultralytics    \
    import YOLO
```

```python
from wandb import (
    init
)

from wandb.integration.ultralytics \
    import (
        add_wandb_callback
)
```

```python
# Setup of environment.
environ['WANDB_DIR'] = '~/wandb'

environ['WANDB_CONSOLE']       = 'off'
environ['WANDB_DISABLE_CODE']  = 'off'

environ['WANDB_NOTEBOOK_NAME'] = 'TrainingOfModel.ipynb'
environ['WANDB_ENTITY']        = 'Designermadsen'
environ['WANDB_SILENT']        = 'True'
```

```python
generator   = None
model       = None
logger      = None
model       = None
```

```python
def generateGenerator() -> None:
    global generator

    if generator is None:
        generator = SystemRandom()
```

```python
# Maximum values
def IntegerZero() -> int:
    global ZERO
    return ZERO

def maximumSizeForInteger() -> int:
    return 2147483647
```

```python
# Setters
def setSeed(
    value: int
) -> None:
    global SEED
    SEED = value

def setModel(
    value
):
    global model
    model = value
```

```python
# Generate
def generateSeed() -> None:
    global          \
        SEED,       \
        generator
    
    generateGenerator()

    setSeed(
        generator.randint(
            IntegerZero(), 
            maximumSizeForInteger()
        )
    )

    return SEED
```

```python
# Getters
def getSeed() -> int:
    global SEED
    
    if isSeedNone():
        return generateSeed()

    return SEED

def isSeedNone() -> bool:
    global SEED
    return SEED is None

def getModel():
    global model
    return model

def getJobType() -> str:
    global JOB_TYPE
    return JOB_TYPE

def getConfiguration() -> dict:
    global CONFIGURATION
    return CONFIGURATION

def getConfigurationKey(
    key: str
) -> str:
    return getConfiguration()[key]

def getDatasetLocation() -> str:
    global DATASET_LOCATION
    return DATASET_LOCATION

def getBatches() -> int:
    global BATCHES
    return BATCHES

def getEpochs() -> int:
    global EPOCHS
    return EPOCHS

def getCosLR() -> float:
    global COS_LR
    return COS_LR

def getBoxParameter() -> float:
    global BOX_PARAMETER
    return BOX_PARAMETER

def getDropOut() -> float:
    global DROPOUT
    return DROPOUT

def getConfidence() -> float:
    global CONFIDENCE
    return CONFIDENCE

def getPatience() -> int:
    global PATIENCE
    return PATIENCE

def getMosiacParameter():
    global MOSAIC_PARAMETER
    return MOSAIC_PARAMETER

def getMosiacEndAt():
    global MOSAIC_END_AT
    return MOSAIC_END_AT

def getImageDimensions():
    global IMAGE_DIMENSIONS
    return IMAGE_DIMENSIONS

def getIsToResume():
    global IS_TO_RESUME
    return IS_TO_RESUME

def getLoadDataSetWorkers():
    global LOAD_DATASET_WORKERS
    return LOAD_DATASET_WORKERS

def getLoadStreamBuffer():
    global LOAD_STREAM_BUFFER
    return LOAD_STREAM_BUFFER

def getDevicesToUse():
    global DEVICES_TO_USE
    return DEVICES_TO_USE
```

```python
def initializeLogging() -> None:
    logger = getLogger(
        __name__
    )

    logger.setLevel(
        ERROR
    )
```

```python
def SetupOfTraining():
    initializeLogging()

    init(
        project     = PROJECT_NAME, 
        job_type    = getJobType(),
        save_code   = False
    )

    setModel(
        YOLO(
            LOAD_MODEL_WEIGHTS_FROM,
            verbose = False
        )
    )

    add_wandb_callback(
        model,
        enable_train_validation_logging     = False,
        enable_validation_logging           = False,
        enable_prediction_logging           = True,
        enable_model_checkpointing          = False
    )
```

```python
def Train():
    SetupOfTraining()
    
    global HAS_RUN

    if not HAS_RUN:
        HAS_RUN = True
    
    model = getModel()

    training_results =                          \
        model.train(
            data         = getDatasetLocation(), 

            batch        = getBatches(), 
            epochs       = getEpochs(),
            cos_lr       = getCosLR(),
            box          = getBoxParameter(),
            iou          = calculateIOU(),
            dropout      = getDropOut(),
            conf         = getConfidence(),
            patience     = getPatience(),
                
            auto_augment = 'randaugment',
            bgr          = 0.0,
            hsv_h        = 0.0,
            hsv_s        = 0.0,
            hsv_v        = 0.0,

            translate    = 0.30,
            scale        = 0.75,
            mixup        = 0.15,
                
            mosaic        = getMosiacParameter(),
            close_mosaic  = getMosiacEndAt(),
                
            imgsz         = getImageDimensions(),
            multi_scale   = False,
            resume        = getIsToResume(),
            seed          = getSeed(),
            workers       = getLoadDataSetWorkers(),
            stream_buffer = getLoadStreamBuffer(),
            device        = getDevicesToUse(),
                
            save          = True,

            plots         = True,
            show          = False,
            verbose       = False
        )
    
    return training_results
```

```python
for idx in range(
    NUMBER_OF_TIMES_TO_ITERATE_TRAINING
):
    CURRENT_NUMBER_OF_TRAINING_ITERATION = NUMBER_OF_TIMES_TO_ITERATE_TRAINING

    if not IS_TO_RESUME or HAS_RUN:
        generateSeed()

    Train()
```
