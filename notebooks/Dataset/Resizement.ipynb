```python
RESIZE_FILES_IN_DIRECTORY: str = '/mnt/c/Users/Kentv/Desktop/Billeder'
RESIZE_TO: int = 1024
```

```python
from os \
    import listdir

from os.path \
    import (
        isdir, 
        isfile, 
        join
)

from math import ceil

from tqdm import tqdm
from PIL import Image
```

```python
found: list = listdir(
    RESIZE_FILES_IN_DIRECTORY
)

for idx in range(
    len(
        found
    )
):
    found[idx] = join(
        RESIZE_FILES_IN_DIRECTORY, 
        found[idx]
    )
```

```python
for idx in tqdm(
    range(
        len(
            found
        )
    )
):
    path: str = found[idx]
    
    with Image.open(
        path
    ) as image:
        width: int = image.width
        height: int = image.height
        
        SKIP: bool =    (RESIZE_TO >= width) \
                        or                   \
                        (RESIZE_TO >= height)

        # Only if its above the resize height
        if not SKIP:
            heighest_number = max(
                width, 
                height
            )

            if(
                heighest_number == width
            ):
                # If it chooses the width as heighest
                new_factor: float = float(
                    width/RESIZE_TO
                )
                
                new_width: int = ceil(
                    width / new_factor
                )

                new_height: int = ceil(
                    height / new_factor
                )

                image = image.resize(
                    (
                        new_width,
                        new_height
                    )
                )

                image.save(
                    path
                )

            else:
                # If it chooses the height as heighest
                new_factor: float = float(
                    height/RESIZE_TO
                )

                new_width: int = ceil(
                    width/new_factor
                )

                new_height: int = ceil(
                    height/new_factor
                )

                image = image.resize(
                    (
                        new_width,
                        new_height
                    )
                )

                image.save(
                    path
                )



    
```
