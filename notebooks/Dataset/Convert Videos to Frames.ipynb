```python
from os import listdir, makedirs
from os.path import isdir, isfile, join, basename

import cv2

DirectoryToLoadVideos: str = '/mnt/c/Users/Kentv/Desktop/Spectre/Videos'
DirectoryToSaveFrames: str = '/media/frames'

counter_frame: int = 1
save_frame_by: int = 1
```

```python
if isdir(DirectoryToSaveFrames):
    %rm {DirectoryToSaveFrames} -R

makedirs(DirectoryToSaveFrames)
```

```python
def setFrameCounter(
    value: int
) -> None:
    global counter_frame
    counter_frame = value

def incrementOfFrameCounter() -> int:
    global counter_frame
    one: int = 1

    setFrameCounter(
        counter_frame 
        + 
        one
    )

    return getFrameCounter()

def getFrameCounter() -> int:
    global counter_frame
    return counter_frame

def resetFrameCounter() -> int:
    global counter_frame
    return counter_frame
```

```python
if not isdir(
    DirectoryToSaveFrames
):
    makedirs(
        DirectoryToSaveFrames
    )
```

```python
found_files: list = list()

for found in listdir(
    DirectoryToLoadVideos
):
    full_path_to_file: str = join(
        DirectoryToLoadVideos, 
        found
    )

    if isfile(
        full_path_to_file
    ):
        found_files.append(
            full_path_to_file
        )
```

```python
def remove_extension(filename):
    from os.path import basename
    zero: int = 0
    one: int = 1

    returnable: str = basename(filename)
    current_position = len(returnable) - one

    while(True):
        current_position = current_position - one

        if returnable[current_position] == '.':
            break

        if current_position == zero:
            break

    return returnable[zero:current_position]
```

```python
from PIL import Image

zero: int = 0

for file in found_files:
    capture = cv2.VideoCapture(
        file
    )
    
    file_name: str = remove_extension(
        file
    )

    while capture.isOpened():
        returnable, frame = capture.read()
        if not returnable:
            break

        full_file_name = join(
            DirectoryToSaveFrames, 
            str(
                file_name 
                + 
                '_' 
                + 
                str(getFrameCounter()) 
                + 
                '.jpg'
            )
        )

        frame = cv2.cvtColor(
            frame, 
            cv2.COLOR_BGR2RGB
        )

        if (
            int(
                getFrameCounter()
                %
                save_frame_by
            ) 
            == 
            zero
        ):
            save_as = Image.fromarray(
                frame
            )

            save_as.save(
                full_file_name,
                subsampling = zero,
                quality     = 100
            )

        incrementOfFrameCounter()
    
    resetFrameCounter()
    capture.release()
```

```python
filter_on_end: bool = True
limit_by: int = 1060
```

```python
from os import listdir
from os.path import isdir

found_files: list = listdir(
    DirectoryToSaveFrames
)

last_position: int = len(found_files)

while not (last_position == 0):
    last_position = last_position - 1
    file = found_files[last_position]
    file_path = join(DirectoryToSaveFrames, file)
    found_files[last_position] = file_path

    if isdir(file_path):
        found_files.pop(last_position)
```

```python
from random import SystemRandom
from os import remove

difference: int = len(found_files) - limit_by
generator = SystemRandom()

if 0 < difference and filter_on_end:
    for idx in range(difference):
        selected_file_index = generator.randint(
            0, 
            len(found_files) - 1
        )

        selected_file = found_files[selected_file_index]

        remove(selected_file)
        found_files.pop(selected_file_index)

        

```
